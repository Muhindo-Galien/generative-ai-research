```python
from langchain.prompts import PromptTemplate

prompt_template_name = PromptTemplate(
    input_variables=["product"],
    template="What is a good name for a company that makes {product}?",
)

new_promt = PromptTemplate.from_template("What is a good name for a company that makes {product}?")

p = new_promt.format(product="colorful socks")
print(p)

```

```python
from langchain.chains import LLMChain
from langchain_community.llms import OpenAI
from langchain_core.prompts import PromptTemplate

llm = LLMChain(llm=OpenAI(), prompt=prompt_template_name, verbose=True)

# print(llm.invoke({"product": "colorful socks"}))
llm.run("colorful socks")
```

# agent and tools

```python
pip install wikipedia
```

```python
from langchain.agents import load_tools, initialize_agent, AgentType
from langchain.llms import OpenAI

llm = OpenAI(temperature=0, api_key=os.environ['OPENAI_API_KEY'])
tools = load_tools(["wikipedia", "llm-math"], llm=llm)
agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True)

agent.run("What was the GDP of the US in 2024?")
```

Memory

```python
from langchain.memory import ConversationBufferMemory

memory = ConversationBufferMemory(memory_key="chat_history")
agent_with_memory = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True, memory=memory)

agent_with_memory.run("What was the GDP of the US in 2024?")
```

```python
agent_with_memory.run("What about the UK in 2024?")
```

```python
print(agent_with_memory.memory.chat_memory.messages)
```
